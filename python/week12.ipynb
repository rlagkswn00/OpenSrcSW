```python
import csv

f = open('seoul_2021.csv',encoding='cp949')
data =csv.reader(f)
max_temp = -999
max_date = 0
min_temp = 1000
min_date = 0
max_min = -999
max_min_date = 0

#최고기온 4
#최저기온 3
#날짜 0
header = next(data)
for row in data:
    #max값
    if float(max_temp) < float(row[4]):
        max_temp = row[4]
        max_date = row[0]
    #min값
    if float(min_temp) > float(row[3]):
        min_temp = row[3]
        min_date = row[0]
    #일교차값
    temp_sub = float(row[4]) - float(row[3])
    if float(max_min) < temp_sub:
        max_min = temp_sub
        max_min_date = row[0]

f.close()
    
#일교차가 가장 큰 날
#날짜
#최저기온이 가장 낮은 날
#날짜
#최고기온이 가장 높은 날
#날짜
print("일교차가 가장 큰 날")
print(max_date)
print("최저기온이 가장 낮은 날")
print(min_date)
print("최고기온이 가장 높은 날")
print(max_min_date)
```

# 실습2

```python
import csv
def show_result(result_dic, line):
    print("이용객이 가장 많은 역 순위 ({0:s})".format(line))
    key = "most_1"
    for i in range(1,4):
        value_list = result_dic.get(key)
        print("{0:d}. {1:s}({2:s}명)".format(i,value_list[0],str(value_list[1])))
        key = key.replace(str(i),str(i+1))
        
    print("이용객이 가장 적은 역 순위 ({0:s})".format(line))
    key = "least_1"
    for i in range(1,4):
        value_list = result_dic.get(key)
        print("{0:d}. {1:s}({2:s}명)".format(i,value_list[0],str(value_list[1])))
        key = key.replace(str(i),str(i+1))
        
f = open('subwayfee_2021.csv', encoding='UTF-8')
data =csv.reader(f)
header = next(data)
#호선 명 1
#지하철 역 3
#승차객수 4
#하차객수 5
result_dic = {'most_1':['',0],'most_2':['',0],
                  'most_3':['',0],'least_1':['',9999999],
                  'least_2':['',9999999],'least_3':['',9999999]}
line = "1호선"
for row in data:
    now_station = row[3]
    now_customer = int(row[4].replace(",",""))+int(row[5].replace(',',""))
    if line == row[1]:
        #최고값 계산
        value_list = result_dic.get("most_1")
        name_1 = value_list[0]
        num_1 = value_list[1]
        value_list2 = result_dic.get("most_2")
        name_2 = value_list2[0]
        num_2 = value_list2[1]
        value_list3 = result_dic.get("most_3")
        name_3 = value_list3[0]
        num_3 = value_list3[1]
        
        if num_1 < now_customer:
            result_dic["most_3"] = result_dic["most_2"]
            result_dic["most_2"] = result_dic["most_1"]
            result_dic["most_1"] = [now_station,now_customer]
            
        elif num_2 < now_customer:
            result_dic["most_3"] = result_dic["most_2"]
            result_dic["most_2"] = [now_station,now_customer]
        
        elif num_3 < now_customer:
            result_dic["most_3"] = [now_station,now_customer]
        
            #최저값 계산
        value_list = result_dic.get("least_1")
        name_1 = value_list[0]
        num_1 = value_list[1]
        value_list2 = result_dic.get("least_2")
        name_2 = value_list2[0]
        num_2 = value_list2[1]
        value_list3 = result_dic.get("least_3")
        name_3 = value_list3[0]
        num_3 = value_list3[1]
        
        if num_1 > now_customer:
            result_dic["least_3"] = result_dic["least_2"]
            result_dic["least_2"] = result_dic["least_1"]
            result_dic["least_1"] = [now_station,now_customer]
            
        elif num_2 > now_customer:
            result_dic["least_3"] = result_dic["least_2"]
            result_dic["least_2"] = [now_station,now_customer]
        
        elif num_3 > now_customer:
            result_dic["least_3"] = [now_station,now_customer]
    else:
        show_result(result_dic,line)
        line = row[1]
        result_dic["most_1"] = ['',0]
        result_dic["most_2"] = ['',0]
        result_dic["most_3"] = ['',0]
        result_dic["least_1"] = ['',9999999]
        result_dic["least_2"] = ['',9999999]
        result_dic["least_3"] = ['',9999999]
        
```
